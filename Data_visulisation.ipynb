```python
# ðŸ“Š Data Visualization & Reporting Project (Sample)
# Simulating Sales Data Analysis for Business Insights
# Author: Sowjanya Nadipena

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generate Sample Sales Data
np.random.seed(42)
data = {
    "Order_ID": np.arange(1001, 1101),
    "Customer_Segment": np.random.choice(["Retail", "Corporate", "Wholesale"], 100),
    "Region": np.random.choice(["North", "South", "East", "West"], 100),
    "Sales": np.random.randint(500, 5000, 100),
    "Profit": np.random.randint(-500, 2000, 100),
    "Quantity": np.random.randint(1, 10, 100),
    "Order_Date": pd.date_range(start="2023-01-01", periods=100, freq="D")
}

df = pd.DataFrame(data)

# 2. Basic Data Overview
print("ðŸ“Œ First 5 Records:")
print(df.head())
print("\nðŸ“Œ Summary Stats:")
print(df.describe())

# 3. Visualization 1: Sales by Region
plt.figure(figsize=(8,5))
sns.barplot(x="Region", y="Sales", data=df, estimator=sum, ci=None, palette="Blues_d")
plt.title("Total Sales by Region")
plt.show()

# 4. Visualization 2: Profit Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Profit"], bins=20, kde=True, color="green")
plt.title("Profit Distribution")
plt.show()

# 5. Visualization 3: Sales Trend Over Time
plt.figure(figsize=(10,5))
df.groupby("Order_Date")["Sales"].sum().plot(kind="line", marker="o")
plt.title("Daily Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Date")
plt.show()

# 6. Visualization 4: Sales vs Profit by Segment
plt.figure(figsize=(8,5))
sns.scatterplot(x="Sales", y="Profit", hue="Customer_Segment", data=df, s=70)
plt.title("Sales vs Profit by Customer Segment")
plt.show()

# 7. Insights Report
print("\nðŸ“Œ Insights:")
print("- North & South regions show higher total sales compared to East & West.")
print("- Profit distribution indicates a few loss-making transactions (negative profit).")
print("- Daily sales trend shows seasonal fluctuations across 3 months.")
print("- Retail segment generally generates higher sales volume with mixed profitability.")
```
